```python
import pandas as pd
```

```python
data = pd.read_csv('./experiment_data_prelims (1).csv')
data.head(20)
```

```python
rows = data['response'] == '{}'
data = data.loc[~rows,:]
data
```

```python
ratings = [int(i) for i in data.iloc[1::2,:]['response'].values]
data = data.iloc[::2,:]
# data = data.iloc[:-1,:]
data['ratings'] = ratings
data
```

```python
data['stim'] = [eval(i)[0].split('.')[0][:-4] for i in data['stim']]
data['stim'] = [i.replace("_", " ") for i in data['stim']]
data['slider1'] = [eval(i)[0] for i in data['resp']]
data['slider2'] = [eval(i)[1] for i in data['resp']]
data['slider1'] = [i.replace("<br>", " ") for i in data['slider1']]
data['slider2'] = [i.replace("<br>", " ") for i in data['slider2']]
data
```

```python
ratings = []
for i,row in data.iterrows():
    if row['stim'] == row['slider1']:
        ratings.append(100-row['ratings'])
    else:
        ratings.append(row['ratings'])
data['ratings_corrected'] = ratings
data
```

```python
data.sort_values(by=['slider1'],inplace=True)
data
```

```python
import seaborn as sns
import numpy as np
```

```python
all_vids = np.unique(data['slider1'])
all_vids
```

```python
mat = np.zeros((all_vids.shape[0],all_vids.shape[0]))
for r,row in enumerate(all_vids):
    for c,col in enumerate(all_vids):
        row1 = (data['slider1'] == row) & (data['slider2'] == col)
        if row == col:
            mat[r,c] = np.nan
        else:
            try:
                mat[r,c] = data.loc[row1,'ratings_corrected'].values[0]
            except:
                 mat[r,c] = np.nan
mat
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(data=mat, annot=True, cmap = 'RdBu_r')
plt.xticks(.5+np.arange(17),all_vids, rotation=90)
plt.yticks(.5+np.arange(17),all_vids, rotation=0)
```
